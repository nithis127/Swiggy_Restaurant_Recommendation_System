## import data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("swiggy.csv")
df
```

## Data Cleaning

```python
df.info()
```

```python
df.set_index("id", inplace=True)
df.drop(["lic_no","link","address","menu"], axis=1, inplace=True)
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
for i in df.columns:
    if df[i].isnull().sum() > 0:
        print(i,":", df[i].isnull().sum())
```

```python
df1 = df.loc[df["name"].isnull()]
df1
```

```python
df1.info()
```

```python
df.dropna(subset=['name'], inplace=True)

for i in df.columns:
    if df[i].isnull().sum() > 0:
        print(i,":", df[i].isnull().sum())
```

```python
df.dropna(subset=['cuisine'], inplace=True)

for i in df.columns:
    if df[i].isnull().sum() > 0:
        print(i,":", df[i].isnull().sum())
```

```python
df["cost"] = df["cost"].ffill()

for i in df.columns:
    if df[i].isnull().sum() > 0:
        print(i,":", df[i].isnull().sum())
```

```python
df
```

```python
df["city"].unique()
```

```python
split_data = df['city'].str.split(',', n=1, expand=True)

# Assign to new columns
df['Area'] = split_data[0].str.strip()
df['City'] = split_data[1].str.strip()

# Fill missing city names with area (when no comma was present)
df['City'] = df['City'].fillna(df['Area'])
df['Area'] = df['Area'].where(df['Area'] != df['City'], None)

# Fill None values in Area with 'Unknown'
df['Area'] = df['Area'].fillna('Unknown')
```

```python
df
```

```python
# Remove ₹ and convert to integer
df['cost'] = df['cost'].str.replace('₹', '', regex=False).str.strip().astype(int)

df.info()
```

```python
df["rating"].unique()
```

```python
df["rating_count"].unique()
```

```python
def clean_rating(x):
    if x == '--':
        # Replace missing rating with a random float between 2.5 and 4.5
        return round(np.random.uniform(2.5, 4.0), 1)
    else:
        return float(x)

def clean_rating_count(x):
    if x == 'Too Few Ratings':
        return np.random.randint(1, 10)
    elif x == '20+ ratings':
        return np.random.randint(20, 50)
    elif x == '50+ ratings':
        return np.random.randint(50, 100)
    elif x == '100+ ratings':
        return np.random.randint(100, 500)
    elif x == '500+ ratings':
        return np.random.randint(500, 1000)
    elif x == '1K+ ratings':
        return np.random.randint(1000, 5000)
    elif x == '5K+ ratings':
        return np.random.randint(5000, 10000)
    elif x == '10K+ ratings':
        return np.random.randint(10000, 20000)
```

```python
df['rating'] = df['rating'].apply(clean_rating)
df['rating_count'] = df['rating_count'].apply(clean_rating_count)
df.info()
```
