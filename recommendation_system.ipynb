```python
def get_restaurants(city, cuisine, top_n = 15):

    # Filter restaurants by city
    city_df = cleaned_df[cleaned_df['City'].str.lower() == city.lower()]

    # Filter restaurants in that city by cuisine
    cuisine_mask = city_df['cuisine'].str.contains(fr'\b{cuisine}\b', case=False, regex=True)
    
    restaurant_ids = city_df[cuisine_mask].index

    # Select encoded features for city and cuisine
    city_encoded = encoded_df.loc[city_df.index]
    cuisine_encoded = encoded_df.loc[restaurant_ids]

   # Compute cosine similarity (within same city only)
    sim_matrix = cosine_similarity(cuisine_encoded, city_encoded)

    # Take the mean similarity for each restaurant in city
    mean_similarity = sim_matrix.mean(axis=0)

    # Sort by similarity (descending)
    sorted_idx = np.argsort(mean_similarity)[::-1]

    # Pick top N restaurants
    similar_restaurants = city_df.iloc[sorted_idx].head(top_n).copy()
#    similar_restaurants['similarity_score'] = mean_similarity[sorted_idx][:top_n]

    if list(city_df["Area"].unique()) == ["Unknown"]:
        return similar_restaurants[["name",	"rating", "rating_count", "cost", "cuisine"]].reset_index(drop=True)
    else:
        return similar_restaurants[["name",	"rating", "rating_count", "cost", "cuisine", "Area"]].reset_index(drop=True)
```

```python
import re

def get_restaurants(city, cuisine, top_n=15):
    # Filter restaurants by city
    city_df = cleaned_df[cleaned_df['City'].str.lower() == city.lower()]

    # Handle multiple cuisines like "Sweets,Bakery"
    cuisine_list = [c.strip() for c in cuisine.split(',')]

    # Build regex to match any cuisine (OR condition)
    pattern = '|'.join([fr'\b{re.escape(c)}\b' for c in cuisine_list])

    # Filter by any of the cuisines
    cuisine_mask = city_df['cuisine'].str.contains(pattern, case=False, regex=True, na=False)
    restaurant_ids = city_df[cuisine_mask].index
    if restaurant_ids.empty:
        # No matching restaurants found — return top rated from city
        return city_df.sort_values('rating', ascending=False).head(top_n)[
            ["name", "rating", "rating_count", "cost", "cuisine", "Area"]
        ].reset_index(drop=True)

    # Select encoded features
    city_encoded = encoded_df.loc[city_df.index]
    cuisine_encoded = encoded_df.loc[restaurant_ids]

    # Compute cosine similarity safely
    sim_matrix = cosine_similarity(cuisine_encoded, city_encoded)
    mean_similarity = sim_matrix.mean(axis=0)

    # Sort and select top N
    sorted_idx = np.argsort(mean_similarity)[::-1]
    similar_restaurants = city_df.iloc[sorted_idx].head(top_n).copy()
    similar_restaurants['similarity_score'] = mean_similarity[sorted_idx][:top_n]

    # Choose columns dynamically
    cols = ["name", "rating", "rating_count", "cost", "cuisine"]
    if list(city_df["Area"].unique()) != ["Unknown"]:
        cols.append("Area")

    return similar_restaurants[cols].reset_index(drop=True)
```

```python
def get_restaurants(city, cuisine, top_n=15):
    # Filter restaurants by city
    city_df = cleaned_df[cleaned_df['City'].str.lower() == city.lower()]

    # Filter by exact cuisines (case-insensitive)
    cuisine_df = city_df[city_df['cuisine'].str.lower().isin([c.lower() for c in cuisine])]

    # Get matching indices
    restaurant_ids = cuisine_df.index

    # Select encoded features
    city_encoded = encoded_df.loc[city_df.index]
    cuisine_encoded = encoded_df.loc[restaurant_ids]

    # Compute cosine similarity safely
    sim_matrix = cosine_similarity(cuisine_encoded, city_encoded)
    mean_similarity = sim_matrix.mean(axis=0)

    # Sort and select top N
    sorted_idx = np.argsort(mean_similarity)[::-1]
    similar_restaurants = city_df.iloc[sorted_idx].head(top_n).copy()
    similar_restaurants['similarity_score'] = mean_similarity[sorted_idx][:top_n]

    # Choose columns dynamically
    cols = ["name", "rating", "rating_count", "cost", "cuisine"]
    if list(city_df["Area"].unique()) != ["Unknown"]:
        cols.append("Area")

    return similar_restaurants[cols].reset_index(drop=True)
```
