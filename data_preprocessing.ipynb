```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer, OneHotEncoder, StandardScaler
```

```python
df = pd.read_csv("cleaned_data.csv", index_col="id")
df
```

```python
df.info()
```

```python
df.drop("name", axis=1, inplace=True)
df
```

```python
# Split the comma-separated strings into lists
df['cuisine'] = df['cuisine'].str.split(',')

# Apply MultiLabelBinarizer
mlb = MultiLabelBinarizer()
cuisine_encoded = mlb.fit_transform(df['cuisine'])

# Convert to DataFrame with proper column names
cuisine_df = pd.DataFrame(cuisine_encoded, columns=mlb.classes_, index=df.index)

# Concatenate with original DataFrame
df = pd.concat([df.drop('cuisine', axis=1), cuisine_df], axis=1)
```

```python
df
```

```python
# Create and fit encoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded = encoder.fit_transform(df[['City', 'Area']])

# Convert encoded data to DataFrame (keep same index)
encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(['City', 'Area']),
    index=df.index
)

# Combine with original DataFrame (drop old columns)
df = pd.concat([df.drop(['city', 'City', 'Area'], axis=1), encoded_df], axis=1)
```

```python
df
```

```python
scaler = StandardScaler()
scaler_cols = ["rating", "rating_count", "cost"]

df[scaler_cols] = scaler.fit_transform(df[scaler_cols])
df
```

```python
df.info()
```

```python
df.to_csv("encoded_data.csv", index=True)
```
